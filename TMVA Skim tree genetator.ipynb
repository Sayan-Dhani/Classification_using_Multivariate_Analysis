```python
eta_max = 2.1    #  this could have two value 2.1 (whole detector) or 1.4(central region)
```

```python
import ctypes

f = root.TFile.Open("sig_tree2_ZMM_s.root","RECREATE")
t1 = root.TTree("tree","tree")


ROOT.gROOT.ProcessLine(
"struct myvar_t {\
    double          dR;\
    double          dPhiMET_u1;\
    double          mass_uu;\
    double          MET_pt;\
    double          pTu1;\
    double          pTu2;\
    int           Nele;\
    int           Njet;\
    int           Nbjet;\
    int           Nmuon;\
    double          HpT_jet;\
 };");


varList = ROOT.myvar_t()

t1.Branch('dR',ROOT.addressof(varList,"dR"),'dR/D')
t1.Branch('dPhiMET_u1',ROOT.addressof(varList,"dPhiMET_u1"),'dPhiMET_u1/D')
t1.Branch('mass_uu',ROOT.addressof(varList,'mass_uu'),'mass_uu/D')
t1.Branch("MET_pt",ROOT.addressof(varList,"MET_pt"),"MET_pt/D")
t1.Branch("pTu1",ROOT.addressof(varList,"pTu1"),"pTu1/D")
t1.Branch("pTu2",ROOT.addressof(varList,"pTu2"),"pTu2/D")
t1.Branch("Nele",ROOT.addressof(varList,"Nele"),"Nele/I")
t1.Branch("Njet",ROOT.addressof(varList,"Njet"),"Njet/I")
t1.Branch("Nbjet",ROOT.addressof(varList,"Nbjet"),"Nbjet/I")
t1.Branch("Nmuon",root.addressof(varList,"Nmuon"),"Nmuon/I")
t1.Branch("HpT_jet",root.addressof(varList,"HpT_jet"),"HpT_jet/D")

start = time.time()    # note the starting time

for i,ent in enumerate(chain):
    
    ### ================================= Event number block ========================================== ###
    if i>5000000: break;
    if (i > 0 and i%1000000 == 0): print("Processing: ", i, "th event!")
    
    ### ================================================================================================ ###
    ### ==================================== MAJOR SELECTION BLOCK ===================================== ###
    ### ================================================================================================ ###
    
    ### ================================= Pre selection of events ================================== ###
    if ent.HLT_IsoMu24_eta2p1 !=1: continue;
    #if ent.MET_pt >40 : continue;
    if ent.nMuon <2 : continue;
    
    # Muon properties block
    u_pt       = list(ent.Muon_corrected_pt)
    sum_pt     = sum(u_pt)
    u_eta      = list(ent.Muon_eta)
    u_phi      = list(ent.Muon_phi)
    u_mass     = list(ent.Muon_mass)
    u_charge   = list(ent.Muon_charge)
    u_dxy      = list(ent.Muon_dxy)
    u_dz       = list(ent.Muon_dz)
    u_iso      = list(ent.Muon_miniPFRelIso_all)
    u_tightId  = list(ent.Muon_tightId)
    u_mediumId = list(ent.Muon_mediumId)
    u_looseId  = list(ent.Muon_looseId)
    
    
    ### ===================================  GOOD Selection Block ======================================= ###
    goodid = []
    for k in range(ent.nMuon):
        if(u_pt[k]> 24 and abs(u_eta[k]) < eta_max and u_iso[k] < 0.3 and u_tightId[k]==1): goodid.append(k)

    # Require at least 1 good muon
    if len(goodid) < 1: continue;
        
    # randomize the goodid list
#     if len(goodid) > 1: random.shuffle(goodid)
    gidx = goodid[0]
        
    # get four vector of the good muon
    u1LV = root.TLorentzVector()
    u1LV.SetPtEtaPhiM(u_pt[gidx],u_eta[gidx],u_phi[gidx],u_mass[gidx])
    varList.pTu1 = u_pt[gidx]
    varList.dPhiMET_u1 = root.TVector2().Phi_mpi_pi(u_phi[gidx] - ent.MET_phi)
#     print(pTu1)
    
 
    ### ===================================  Medium Selection Block ===================================== ###
    midid = []
    for k in range(ent.nMuon):
        if k == gidx: continue
        if (u_charge[gidx]*u_charge[k] > 0): continue;
        if abs(u_eta[k]) > eta_max: continue
        # soft cut on pt and isolation
        if (u_pt[k] < 5):continue
        if (u_iso[k] > 0.4) : continue
        if u_mediumId[k] == 1: midid.append(k)
    if len(midid) < 1: continue
#     if len(midid) > 2:continue
    varList.Nmuon = len(midid)+1

    massdif = 999.0
    sindx = -1
    mz = 91.2
    for k in midid:
        u2LV = root.TLorentzVector()
        u2LV.SetPtEtaPhiM(u_pt[k],u_eta[k],u_phi[k],u_mass[k])

        Z_LV = u1LV + u2LV
        mass = Z_LV.M()
        if 20 < mass < 160:
            if abs(mass - mz) < massdif: 
                massdif = abs(mass - mz)
                sindx = k
            
    if sindx == -1: continue            
    u2LV = root.TLorentzVector()
    u2LV.SetPtEtaPhiM(u_pt[sindx],u_eta[sindx],u_phi[sindx],u_mass[sindx])    
#     varList.mass_uu = invmass
    varList.dR = u1LV.DeltaR(u2LV)
    varList.mass_uu = (u1LV+u2LV).M()
    varList.pTu2 = u_pt[sindx]
    
    
    ### =================================== Electron Veto Block ========================================== ###
    ele_pt    = list(ent.Electron_pt)
    ele_eta   = list(ent.Electron_eta)
    ele_phi   = list(ent.Electron_phi)
    ele_mass  = list(ent.Electron_mass)
    ele_mvaid = list(ent.Electron_mvaFall17V2Iso_WPL)
    
    eleid = []
    for j in range(ent.nElectron):
        if ele_pt[j]<10: continue
        if abs(ele_eta[j]) > 2.5: continue
        if ele_mvaid[j]==0: continue
        ele_LV = root.TLorentzVector()
        ele_LV.SetPtEtaPhiM(ele_pt[j], ele_eta[j], ele_phi[j], ele_mass[j])            
        if (u1LV.DeltaR(ele_LV)<0.4 or u2LV.DeltaR(ele_LV)<0.4): continue
        eleid.append(j)        
    varList.Nele = len(eleid)
#     print(Nele)
    
    ### ================================= Jet & bJet Veto Block ========================================== ###
    jet_pt   = list(ent.Jet_pt)
    jet_eta  = list(ent.Jet_eta)
    jet_phi  = list(ent.Jet_phi)
    jet_mass = list(ent.Jet_mass)
    jet_btag = list(ent.Jet_btagDeepFlavB)
    jetid  = []
    bjetid = []
    for j in range(ent.nJet):
        if jet_pt[j] < 30: continue
        if abs(jet_eta[j])> 4.7: continue # this is based on the calorimeter
        jet_LV = root.TLorentzVector()
        jet_LV.SetPtEtaPhiM(jet_pt[j], jet_eta[j], jet_phi[j], jet_mass[j])
        if (u1LV.DeltaR(jet_LV)<0.4 or u2LV.DeltaR(jet_LV)<0.4): continue
        
        jetid.append(j)

        if abs(jet_eta[j]) > 2.5: continue 
        if jet_btag[j] <= 0.33: continue
        bjetid.append(j)

    varList.Njet = len(jetid)
#     print(Njet)
    varList.Nbjet = len(bjetid)
#     print(Nbjet)
    if len(jetid) == 0:
        varList.HpT_jet = 0.0
    else:
        varList.HpT_jet = jet_pt[jetid[0]]
#         print(HpT_bjet)
    
    varList.MET_pt = ent.MET_pt


    t1.Fill()
    
print ("Loop is completed")     # LOOP COMPLETTION MASSAGE

t1.Write()
print(t1)
f.Close()
end = time.time()   # Note the end time
print(end-start)    # Show the time taken in seconds to rum the code
```
